```python
from datetime import datetime
from matplotlib import pyplot as plt

lines: list = open('./cameraDetections.txt').read().splitlines()

timestamps: list = []

for line in lines:
    if '2021-' in line:
        continue
    if '+04' in line:
        line = line.replace('+04', '')
    if '.' not in line:
        line += '.0'
    timestamps.append(datetime.strptime(line, '%Y-%m-%d %H:%M:%S.%f').timestamp())

print(timestamps)
```
