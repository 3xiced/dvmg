```python
# Test on real data
from datetime import datetime, timedelta
from matplotlib import pyplot as plt

import sys
import os

sys.path.insert(0, os.path.abspath('..'))

from dvmg.utils import *
from collections import Counter
import numpy as np

lines: list = open('../cameraDetections.txt').read().splitlines()

timestamps: list = list()

for line in lines:
    if '2021-' in line:
        continue
    if '+04' in line:
        line = line.replace('+04', '')
    if '.' not in line:
        line += '.0'
    time = datetime.strptime(line, '%Y-%m-%d %H:%M:%S.%f') - timedelta(hours=1)
    timestamps.append(time.timestamp())

dt_strings: list[str] = list()
dt_objects: list[datetime] = list()
for tmsp in timestamps:
    dt_strings += [datetime.fromtimestamp(
        tmsp).strftime('%Y-%m-%d %H:%M:%S.%f')]
    dt_objects += [datetime.fromtimestamp(tmsp)]

DEVIDER = 20

counter = 0
start_time = dt_objects[0] - timedelta(hours=6)
end_time = start_time + timedelta(hours=12)
while end_time < dt_objects[-1]:
    counter += 1
    start_time = start_time + timedelta(hours=6)
    end_time = start_time + timedelta(hours=12)

    to_inspect_timestamps: list[float] = [
        dt.timestamp() for dt in dt_objects if dt > start_time and dt <= end_time]
    to_inspect_timestamps_dt: list[str] = [dt.strftime(
        '%Y-%m-%d %H:%M:%S.%f') for dt in dt_objects if dt > start_time and dt <= end_time]

    output: dict = dict()
    cut_timestamp = [to_inspect_timestamps[i] -
                     min(to_inspect_timestamps) + .001 for i in range(len(to_inspect_timestamps))]
    if len(cut_timestamp) == 0:
        continue

    for time in cut_timestamp:
        output[time] = 1

    interval_length = max(output.keys())/DEVIDER
    events_number = len(output.keys())
    
    if events_number < 30:
        print(f'{events_number} impossible to restore')
        continue

    to_hist = process_coordinates_to_histogram(list(output.keys()), DEVIDER)

    print('Before restoring')

    plt.close('all')
    plt.rcParams["figure.figsize"] = [20, 2]
    plt.bar(list(output.keys()), list(
        output.values()), width=50)  # type: ignore
    plt.yticks([1])
    plt.xlabel("Реальные данные")
    plt.ylabel("событие")
    plt.show()

    plt.close('all')
    plt.rcParams["figure.figsize"] = [20, 2]
    plt.hist(to_hist)  # type: ignore
    plt.yticks([1])
    plt.xlabel("Реальные данные")
    plt.ylabel("Счетчик")
    plt.show()

    print('=================================================')

    repeated_elements = dict(Counter(to_hist))

    to_restore = 1000 - len(output.keys())

    print(repeated_elements)

    output: dict = dict()

    for interval_number in repeated_elements:
        percentage_of_events_to_fill = repeated_elements[interval_number]/events_number
        events_to_generate = round(to_restore * percentage_of_events_to_fill)
        left_border = interval_length * interval_number
        right_border = left_border + interval_length

        rng = np.random.uniform(low=2*left_border if interval_number not in list(repeated_elements.keys())[:3] else left_border, high=2*right_border, size=(events_to_generate,))
        rng = rng.tolist()

        cut_timestamp += rng

    for time in cut_timestamp:
        output[time] = 1

    # print(list(output.keys()))
    to_hist = process_coordinates_to_histogram(list(output.keys()), DEVIDER)

    print(len(output.keys()))
    # print(to_hist)
        
    print('After restoring')
    plt.close('all')
    plt.rcParams["figure.figsize"] = [20, 2]
    plt.bar(list(output.keys()), list(
        output.values()), width=10)  # type: ignore
    plt.yticks([1])
    plt.xlabel("Восстановленные данные")
    plt.ylabel("событие")
    plt.show()

    plt.close('all')
    plt.rcParams["figure.figsize"] = [20, 2]
    plt.hist(to_hist)  # type: ignore
    plt.yticks([1])
    plt.xlabel("Восстановленные данные")
    plt.ylabel("Счетчик")
    plt.show()

    print('=================================================')

    if counter == 10:
        break
```
